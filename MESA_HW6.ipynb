```python
%matplotlib inline

import matplotlib.pyplot as plt
import mesa_reader as mr

l1 = mr.MesaLogDir('/Users/hshishir/Desktop/mesa_work/HW6/LOGS')

p1 = l1.profile_data()

l2 = mr.MesaLogDir('/Users/hshishir/Desktop/mesa_work/HW6_2/LOGS')

p2 = l2.profile_data()

plt.loglog(p1.r_div_R, 10**p1.logRho, label='ZAMS')
plt.loglog(p2.r_div_R, 10**p2.logRho, label='He burning phase')

plt.xlabel("Radius (r/R)")
plt.ylabel("Density (g cm^-3)")
plt.title("Figure 1: Density as a Function of Radius for ZAMS and He Burning Phase (log plot)")

plt.legend()

plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import mesa_reader as mr

l1 = mr.MesaLogDir('/Users/hshishir/Desktop/mesa_work/HW6/LOGS')

p1 = l1.profile_data()

l2 = mr.MesaLogDir('/Users/hshishir/Desktop/mesa_work/HW6_2/LOGS')

p2 = l2.profile_data()

p_rad_by_p1 = (p1.prad)/(p1.pressure)
p_rad_by_p2 = (p2.prad)/(p2.pressure)

plt.plot(p1.r_div_R, p_rad_by_p1, label='ZAMS')
plt.plot(p2.r_div_R, p_rad_by_p2, label='He burning phase')

plt.xlabel("Radius (r/R)")
plt.ylabel("P_rad/P")
plt.title("Figure 6: Radiation Pressure as a Function of Radius for ZAMS and He Burning Phase")

plt.legend()

plt.show()
```
