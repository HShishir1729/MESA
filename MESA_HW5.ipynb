```python
# Enable inline plotting
%matplotlib inline

# Import necessary libraries
import matplotlib.pyplot as plt
import mesa_reader as mr
import numpy as np

# Load and plot data
h = mr.MesaData('/Users/hshishir/Desktop/mesa_work/HW5/LOGS/history.data')

# Create the main plot for the H-R diagram
fig, ax1 = plt.subplots()

# Plot the H-R diagram (Teff vs Luminosity)
ax1.plot(10**h.log_Teff, 10**h.log_L)

# Set axis labels for Teff and Luminosity
ax1.set_xlabel('Effective Temperature (K)')
ax1.set_ylabel('Luminosity (L/Lsun)')
ax1.invert_xaxis()  # Invert the x-axis for Teff (as is standard in H-R diagrams)

# Add spectral types on the secondary x-axis
# Spectral types and their approximate temperature ranges (in Kelvin)
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temp_ranges = [30000, 10000, 7500, 6000, 5200, 3700, 2400]  # Approximate Teff for each spectral type

# Create a secondary x-axis for the spectral types
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())  # Set the limits of the secondary x-axis to match the primary axis
ax2.set_xticks(temp_ranges)
ax2.set_xticklabels(spectral_types)
ax2.set_xlabel('Spectral Type')

# Show the plot
plt.show()
```

```python
# Enable inline plotting
%matplotlib inline

# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import mesa_reader as mr

# Load data from the history file
h = mr.MesaData('/Users/hshishir/Desktop/mesa_work/HW5/LOGS/history.data')

# Convert log_Teff to B-V color index using an approximation
log_Teff = h.log_Teff
B_V = 0.93 * log_Teff - 3.92

# Convert log_L (luminosity) to absolute magnitude M_V
log_L = h.log_L
M_sun = 4.83  # Absolute magnitude of the Sun
M_V = M_sun - 2.5 * log_L

# Create a plot with B-V color index and absolute magnitude M_V
plt.plot(B_V, M_V, label="Evolutionary Track", color="blue")

# Set axis labels
plt.xlabel('B-V Color Index')
plt.ylabel('Absolute Magnitude $M_V$')

# Invert axes: B-V increases to the right, M_V increases downward
plt.gca().invert_yaxis()  # Brighter stars are at the top
plt.gca().invert_xaxis()  # Hotter stars are to the left (lower B-V)

# Add arrows to indicate directions of increasing B-V and M_V
plt.annotate('Increasing B-V', xy=(0.5, 2), xytext=(1.5, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Increasing M_V', xy=(0.5, 2), xytext=(0.5, 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import mesa_reader as mr
from matplotlib import cm

# Load the history data
h = mr.MesaData('/Users/hshishir/Desktop/mesa_work/HW5/LOGS/history.data')

# Extract the relevant data
log_Teff = h.log_Teff
log_L = h.log_L
star_age = h.star_age  # Star's age for color bar
log_R = h.log_R  # Radius in solar units

# Create a scatter plot for the H-R diagram with color representing time (star_age)
plt.scatter(log_Teff, log_L, c=star_age, cmap=cm.plasma, s=10, edgecolor='none')

# Add a color bar for the time evolution
cbar = plt.colorbar()
cbar.set_label('Star Age (years)')

# Set axis labels
plt.xlabel('log Effective Temperature (K)')
plt.ylabel('log Luminosity (L/Lsun)')

# Add contour lines for log10(R/Rsun)
contour = plt.tricontour(log_Teff, log_L, log_R, levels=10, colors='black', alpha=0.6)
plt.clabel(contour, fmt='%.1f', colors='black')

# Set the x and y limits to include some margin
plt.xlim(min(log_Teff) - 0.05, max(log_Teff) + 0.05)
plt.ylim(min(log_L) - 0.05, max(log_L) + 0.05)

plt.gca().invert_xaxis() 

# Adjust layout to avoid cutting off labels
plt.tight_layout()

# Show the plot
plt.show()
```
